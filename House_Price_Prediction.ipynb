```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dtaset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a pandas dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
house_price_dataframe.head()
```

```python
# add the target column to the dataframe
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# cheking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
# statical measures of the daytaset
house_price_dataframe.describe()
```

Understandin the correlation between various features in the dataset

```python
correlation = house_price_dataframe.corr()
```

```python
# cunstructing a heatmap to measured the correlation
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt= '.2f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and Target

```python
x = house_price_dataframe.drop(['price'], axis =1)
y = house_price_dataframe['price']
```

```python
print(x)
print(y)
```

Splitting the data into Training data and Test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 2 )
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Training (XGBoost regressor)

```python
# loading the model
model = XGBRegressor()
```

```python
# trainig the model with x_train
model.fit(x_train, y_train)
```

Evaluation

Prediction on trainig data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(x_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)

# mean absolute error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)

print("R squared error: ", score_1)
print("Mean Absolute Error :", score_2)
```

Visualizing the actual prices and predicted prices

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()

```

Prediction on test data

```python
test_data_prediction = model.predict(x_test)
```

```python
print(test_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(y_test, test_data_prediction)

# mean absolute error
score_2 = metrics.mean_absolute_error(y_test, test_data_prediction)

print("R squared error: ", score_1)
print("Mean Absolute Error :", score_2)
```

```python
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
